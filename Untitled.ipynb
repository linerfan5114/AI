```python
address=set(df.Address)
Address={}
for i,j in enumerate(address,start=1):
    x={j:i}
    Address.update(x)
number_address=df.Address.map(lambda s:Address.get(s) if s in Address else s)
number_address
```

```python
address=set(df.Address)
Address={}
for i in address:
    j=df[df['Address']==i].Price.sum()/df[df['Address']==i].Area.sum()
    x={i:j}
    Address.update(x)
average_price=df.Address.map(lambda s:Address.get(s) if s in Address else  s)
average_price
```
